## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master

"""
df = pd.read_sql_query(QUERY,conn)

def plot_histogram(data, column, bins=30, title=None, xlabel=None, ylabel='Frequency', color='skyblue', edge_color='black'):
   
    # Set up figure 
    fig, ax = plt.subplots()
    fig.patch.set_alpha(0.0)  

    # Plot histogram
    ax.hist(data[column], bins=bins, color=color, edgecolor=edge_color)

    # Set title and labels
    ax.set_title(f'Histogram of {title}' if title  else f'Histogram of {column}', fontsize=14, fontweight='bold')
    ax.set_xlabel(xlabel if xlabel else column, fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)

    # Style grid and layout
    ax.grid(visible=True, color='grey', linestyle='--', linewidth=0.5)
    plt.tight_layout()  

    # Show plot
    plt.show()
```

```python
plot_histogram(df, 'ConvertedComp')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master

"""
df = pd.read_sql_query(QUERY,conn)
fig = plt.subplots(figsize=(10, 6))

sns.regplot(data=df, x='Age', y='WorkWeekHrs', line_kws={'color': 'red'})

axes.set_title('Regression Plot of Age and Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master

"""
df = pd.read_sql_query(QUERY,conn)
fig, axes = plt.subplots(figsize=(10, 6))
# Create the bubble plot (scatter plot with bubble sizes)
sns.scatterplot(data=df, x='CodeRevHrs', y='WorkWeekHrs', size='Age', sizes=(20, 200), ax=axes)

# Set the title of the plot
axes.set_title('Bubble Plot of Work Week Hours vs Code Review Hours with Age as Bubble Size')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *, COUNT(*) AS count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

data = df['count']
labels = df['DatabaseDesireNextYear']

plt.rcParams['figure.figsize'] = [8,8]

fig, ax = plt.subplots()
ax.pie(data, labels = labels, autopct='%1.1f%%', explode = (0, 0, 0, 0, 0.1), shadow = True)
plt.show()
```

```python
# 2. In the list of most popular languages respondents wish to learn next year, what is the rank of Python?

QUERY = """
SELECT *, COUNT(*) AS count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

df['Rank'] = df['count'].rank(method='dense', ascending=False).astype(int)
python_rank = df[df['LanguageDesireNextYear'] == 'Python']['Rank'].values[0]
table = df[['LanguageDesireNextYear', 'count', 'Rank']].head(5)
print("Top 5 Languages and Their Ranks:")
print(table)

print(f"\nThe rank of Python is: {python_rank}")
```

```python
#  3. How many respondents indicated that they currently work with 'SQL'? 
QUERY = """
SELECT *, COUNT(*) AS count
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY,conn)

sql_count = df[df['LanguageWorkedWith'] == 'SQL']['count'].values[0]
data = df['count'].head(5)
labels = df['LanguageWorkedWith'].head(5)
table = pd.DataFrame({'Labels': labels, 'Data': data})
print("Top 5 Languages:")
print(table)

# Print the count of respondents who work with SQL
print(f"\nThe number of respondents who currently work with 'SQL' is: {sql_count}")
```

```python
# How many respondents indicated that they work on 'MySQL' only? 
QUERY = """
SELECT *
FROM DatabaseWorkedWith

"""
df = pd.read_sql_query(QUERY,conn)
resp_sum = df.groupby('Respondent').sum()
resp_sum[resp_sum['DatabaseWorkedWith']=='MySQL'].count()
```

```python
# What developer type are the majority of the survey responders?
QUERY = """
SELECT DevType, COUNT(*) as Count
FROM DevType
GROUP BY DevType
ORDER BY Count DESC

"""
pd.read_sql_query(QUERY,conn)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)

# Group by Age and calculate median
grouped = df.groupby('Age').median()

# Extract medians for WorkWeekHrs and CodeRevHrs
work_week_medians = grouped['WorkWeekHrs']
code_rev_medians = grouped['CodeRevHrs']

# Create a stacked bar chart
plt.rcParams['figure.figsize'] = [10, 6]

fig, ax = plt.subplots()

# Plot stacked bars for each age
ax.bar(work_week_medians.index, work_week_medians, label='WorkWeekHrs', color='b')
ax.bar(work_week_medians.index, code_rev_medians, bottom=work_week_medians, label='CodeRevHrs', color='g')

# Customize chart
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_ylabel('Hours')
ax.set_xlabel('Age')
ax.set_title('Work Week Hours and Code Review Hours for Age 30-35')

plt.tight_layout()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY, conn)

median_converted_comp = df.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 6))
plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o', linestyle='-', color='b')

plt.xlabel('Age')
plt.ylabel('Median Compensation')
plt.title('Median Compensation for Ages 45 to 60')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Question 2: Based on the Line Chart of Age and median ConvertedComp for the age group 25-30, what kind of correlation do you see between Age and median ConvertedComp? 
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 25 AND 30
"""
df = pd.read_sql_query(QUERY, conn)

median_converted_comp = df.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 6))
plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o', linestyle='-', color='b')

plt.xlabel('Age')
plt.ylabel('Median Compensation')
plt.title('Median Compensation for Ages 25 to 30')

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

mainbranch_counts = df['MainBranch'].value_counts().reset_index()
mainbranch_counts.columns = ['MainBranch', 'Count']

plt.figure(figsize=(14, 8))
sns.barplot(data=mainbranch_counts, y='MainBranch', x='Count',
            hue='MainBranch', palette=['blue', 'red'], legend=False)

plt.xlabel('Total Number')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branch')

plt.tight_layout()
plt.show()
```
